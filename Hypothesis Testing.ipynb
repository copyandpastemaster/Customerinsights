```python
import warnings
warnings.filterwarnings("ignore")

# loading packages
# basic + dates 
import numpy as np
import pandas as pd
from pandas import datetime


# importing train data to learn
train = pd.read_csv("god.csv", 
                    parse_dates = True, low_memory = False, index_col = 'date')
train.index

# data extraction
train['Year'] = train.index.year
train['Month'] = train.index.month
train['Day'] = train.index.day
train['WeekOfYear'] = train.index.weekofyear

# adding new variable
train['sale_per_customer'] = train['sales']/train['customers']
```

```python
sundaya = train[train.brand == "a"].loc[:, ["sales"]]
sundayanot = train[(train.brand == "a") & (train.day_of_the_week != 7)].loc[:, ["sales"]]
```

```python
from scipy.stats import ttest_ind
import numpy as np
sunday = sundaya
sundaynot = sundayanot
print(sunday)
print("Not Sunday Data :-\n")
print(sundaynot)
sunday_mean = np.mean(sunday)
sundaynot_mean = np.mean(sundaynot)
print("Sunday mean value:",sunday_mean)
print("Not Sunday mean value:",sundaynot_mean)
sunday_std = np.std(sunday)
sundaynot_std = np.std(sundaynot)
print("Sunday std value:",sunday_std)
print("Not Sunday std value:",sundaynot_std)
ttest,pval = ttest_ind(sunday,sundaynot)
print("p-value",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```
